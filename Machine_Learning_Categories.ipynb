# Machine Learning Categories

In machine learning there are four differnet types of supervision categories:

    1 - Supervised Learning
    2 - Unsupervised Learning
    3 - Semisupervised Learning
    4 - Reinforcement Learning

## Supervised Learning

In this category the training data has labels. Like the image below the items are fruits and each fruit is visible to us which one is apple or not. The famous *supervised learning* is *classification* such as spam filtering. 



The most important *supervised learning* algorithms are:

    I- k-Nearest Neighbors
    II- Linear Regression
    III- Logistic Regression
    IV- Suport Vector Machines (SVMs)
    V- Decision Trees and Random Forests
    VI- Neural networks

## Unsupervised Learning 

The training data is unlabeled. As you see from the image below it is hard to distinguish between the training dataset as they don't have any labels to be separated. We want to find the similarities between the unlabeled data such that we can group them together. 



The famous algorithms in *Unsupervised Learning* are:

    * Clustering
        1- K-Means
        2- DBSCAN
        3- Hierarchical Cluster Analysis (HCA)
        
    * Anomoly detection and novelty detection
        1- One-class SVM
        2- Isolation Forest
        
    * Visuzlization and dimensionality reduction
        1- Principal Component Analysis (PCA)
        2- Kernel PCA
        3- Locally-Linear Embedding (LLE)
        4- t-distributed Stochastic Neighbor Embedding (t-SNE)
     
     * Association rule learning
        1- Apriori
        2- Eclat

## Semisupervised Learning

If the most portion of the training data is unlabeled and small portion is labeled then we are dealing with *Semisupervised Learning*.



## Reinforement Learning

The learning system studies the environment. The learning system is called *agent*. The agent does two jobs: selection and action. If the agent does the job correctly gets positive rewards and if it doesn't then it receives negative rewards. The agent choose the best approach to get the most rewards. The approach here is called *policy*.
